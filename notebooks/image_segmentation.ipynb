```python
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
img = cv2.imread("butterfly.png")
orignal_shape = img.shape
orignal_shape 
```

```python
plt.imshow(img)
plt.show()
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python
all_pixels = img.reshape((-1, 3))
```

```python
dominant_color = 4
km = KMeans( n_clusters=dominant_color)
```

```python
km.fit(all_pixels)
```

```python
center = np.array(km.cluster_centers_, dtype ='uint8')
```

```python
i = 1
colors =[]
plt.figure(0,figsize =(4,2))
for each_col in center:
    plt.subplot(1,4,i)
    plt.axis("off")
    i+=1
    colors.append(each_col)
    a = np.zeros((100,100,3),dtype='uint8')
    a[:,:,:] = each_col
    plt.imshow(a)
plt.show()
```

```python
new_img = np.zeros((211*319,3), dtype='uint8')
for ix in range(new_img.shape[0]):
    new_img[ix] = colors[km.labels_[ix]]
new_img = new_img.reshape((orignal_shape))
plt.imshow(new_img)
plt.show()
```
